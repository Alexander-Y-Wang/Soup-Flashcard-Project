```python
import pandas as pd
import requests
import urllib
import re
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pickle
import sys
sys.setrecursionlimit(8000)
URL = 'https://www.vocabulary.com/lists/156644'

def main():
    webpage = urlopen(URL)
    soup = BeautifulSoup(webpage, "html.parser")
    collected_words = dict()
    make_words(soup, collected_words)
    
    
def make_words(soup, collected_words):
    words = []
    definitions = []
    word_data = soup.find_all("li", {"word": True})
    for item in word_data:
        item.attrs
        words += item['word']
    for entry in soup.find_all("div", {"class": "definition"}):
        entry.get_text()

    print(words)
    print(definitions)
    
    #Why do these list each character, but print the intended text
    
main()
```
